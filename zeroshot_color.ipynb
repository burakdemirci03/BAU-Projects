```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python
color_dict = {"black": "⚫️", "red": "🔴", "brown": "🟤", "orange": "🟠",
              "yellow": "🟡", "green": "🟢", "blue": "🔵",
              "purple": "🟣", "pink": "🟣+⚪️", "white": "⚪️"}
```

```python
# DistilBERT model for "Zero-Shot Classification" task
model_name = "typeform/distilbert-base-uncased-mnli"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
zero_shot_classifier = pipeline("zero-shot-classification", model=model, tokenizer=tokenizer, multi_label=True)
```

```python
# Function to classify colors in a given sentence
def color_classifier(sentence):
  result = zero_shot_classifier(sentence.strip(" "), candidate_labels=list(color_dict.keys()))
  scores = zip(result["labels"], result["scores"])

  confidence = [(color_dict[label], score) for label, score in scores]

  print("\n")

  template = "{} (%{:.2f})"
  for item in confidence:
    format = template.format(item[0], 100 * item[1])
    if (item[1] > 0.8):
      print(f"{format} ★★★")
    elif (item[1] > 0.5):
      print(f"{format} ★★")
    elif (item[1] > 0.1):
      print(f"{format} ★")
    else:
      print(f"{format}")

  print("\n")
```

```python
# Infinite loop for color classifier
while True:
  sentence = input("Please enter a sentence: ")
  if sentence is None or sentence == "":
    print("*-*-* Terminated loop successfully. *-*-*")
    break
  color_classifier(sentence)
```
